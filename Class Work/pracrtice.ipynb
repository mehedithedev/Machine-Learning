```python
# Using the test_split funciton to split the dataset into training and testing datasets
from sklearn.model_selection import train_test_split # splits the dataset into traiing and testing sets
from sklearn.preprocessing import StandardScaler # standardizes the features by removing the mean and scaling them to unit variance
from sklearn.ensemble import RandomForestClassifier # a machine learning model that builds multiple decesion trees and merges them together to get a more accurate and stable prediciotn
```

```python
# Loading the dataframe: 

df = pd.read_csv("heart-disease.csv")
df.shape
```

```python
# Preprocessing the data: 

X = df.drop('target', axis=1) # dropping the target variable from axis 1 and having X as Capital is a conveintion while y should be lowercase
# X is containing all the patient attributes or features except target
X
```

```python
y = df['target'] # containing the target variable which has been dropped on X
# contaings the outcome we want to predict
y
```

```python
# Splittng the data into 4 variables where
# X_train is a subset of X containing 80% of the data for training 
# X_test is a subset of X containing 20% of the data for testing
# y_train is a subset of y containing 80% of the data for training 
# y_test is a subset of y containing 20% of the data for testing 
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=56)
```
