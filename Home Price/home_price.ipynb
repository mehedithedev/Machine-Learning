```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price($US)')
plt.scatter(df.area, df.price, color="blue", marker="+")
```

```python
reg = linear_model.LinearRegression()
```

```python
# Fitting the data means I'm training the linear regression model using the availbale data 

reg.fit(df[['area']], df.price)
```
