```python
import numpy as np
import matplotlib.pyplot as plt


# Generate some random data points

np.random.seed(0)

# rand() give you uniformly distributed pseudo-random numbers
x = np.random.rand(50) * 10

print("x:", x)

# normal distribution = gaussian or bell-like curve
y = 2 * x + 1 + np.random.randn(50) * 2

print("y:", y)

# Use NumpPy's polyfit method to pefrom linear regression on given data (x,y)

# third argument which we set to "1", indicating that we want to fit a polynomial in the degree 1

slope ,intercept = np.polyfit(x, y, 1)

# Plot data points and regression line
plt.scatter(x, y, label="Data Points")
plt.plot(x, slope * x + intercept, color="r", label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.legend()
plt.grid(True)
plt.show()

print("Slope:", slope)
print("Y-Intercept", intercept)
```

```python
# Geenerate a new x value for prediction
new_x = 7

m = slope
b = intercept
# Predit the corresponding y value using the best fit line equation (y = mx +b)
predicted_y = m * new_x + b
print("for x=", new_x, "the predicted y-value is: ", predicted_y)

# Plot the original date points and the best fit
plt.scatter(x, y, label="Data Points")
plt.plot(x, m *x + b, color="r", label="Best Fit Line")

# Plot the predicted point
plt.scatter(new_x, predicted_y, color="green", label="Predicted Point") 

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.legend()
plt.grid(True)
plt.show()

print("Slope:", slope)
print("Y-Intercept", intercept)
```
